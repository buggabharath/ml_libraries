```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
dataset = np.random.randn(25)
```

```python
sns.rugplot(dataset);
```

```python
x_min = dataset.min() - 2
x_max = dataset.max() + 2
```

```python
x_axis = np.linspace(x_min, x_max, 100)
```

```python
 band = ((4*dataset.std()**5)/(3*len(dataset)))**.2
```

```python
kernal_list =[]
```

```python
for data_point in dataset:
    kernel = stats.norm(data_point, band).pdf(x_axis)
    kernal_list.append(kernel)
    
    
    kernel = kernel / kernel.max()
    kernel = kernel * .4
    plt.plot(x_axis, kernel, color = 'grey', alpha=0.5)
    
```
